### 2.1. Tratamento do ruido via alisamento exponencial simples

$X(t), t = 1, ..., N$\
\
$S(t) = αX(t) + (1 − α)S(t − 1), t = 2, ..., N$\
\
$S(1) = X(1)$

$MAE = \frac{1}{N}\sum_{i=1}^{N} |S_i - X_i|$


```python
def Get_MAE(list_st_alis, Sens):
    
    list_MAE_st = []
    for i in range(0, np.shape(list_st_alis)[0]):
        soma_erro = 0
        for j in range(0 , np.shape(list_st_alis)[1]):
            soma_erro += abs(list_st_alis[i][j] - Sens[:,i][j]) 
        MAE = (1/len(list_st_alis)) * soma_erro 
        list_MAE_st.append(MAE)

    return list_MAE_st
```

```python
#CÓDIGO PYTHON PARA GERAÇÃO DAS TABELAS 1 E 2

# vetor com 10 valores aleatórios do parâmetro alpha
alpha = [0.100, 0.200, 0.300, 0.400, 0.500,
        0.600, 0.700, 0.800, 0.900, 0.999] 

X = Sens # utilizando o dataset completo (treino + teste) para os valores de X (temperatura)
N = X.shape[0] # tamanho da série (quantidade de leituras por 5 dias X 24h X 5 min)
num_ST = X.shape[1] # número de séries temporais (sensores)

list_st_alis_alpha = []
list_MAE_st_alpha = []

for k in range (0,len(alpha)): # laço de 0 a 10 (número de alphas)
    list_st_alis = []
    for j in range (0,num_ST): #laço de 0 a 51 (quantidade de séries)    
        st_alis = []
        for i in range(1,N + 1): # laço de 0 a 14400 (tamanho da série)                  
            if i == 1:
                st = X[0][0] # série temporal alisada com t = 1 / S(1) = X(1)
            else:
                st = alpha[k] * X[i-1][j] + (1-alpha[k]) * st # série temporal alisada com t = 2..N
            
            # lista [14400,1] com todos os valores de X de uma série temporal alisada 
            st_alis.append(st)
            
        # lista [14400,52]  com todos os valores de X de todas as ST alisadas
        list_st_alis.append(st_alis) 
    
    # lista [10,52,14400] com todos os valores de X de todas as séries temporais de cada alpha 
    list_st_alis_alpha.append(list_st_alis) 
        
    # lista [52,1] com os valores de erro MAE de uma ST alisada retornada da função MAE_ST
    list_MAE_st = Get_MAE(list_st_alis, X)
    
    #lista [10,52] com todos os valores de erro MAE de todas as séries temporais
    list_MAE_st_alpha.append(list_MAE_st)
    
#Transporta da lista MAE para o formato [52,10]
list_MAE_st_alpha_T = np.array(list_MAE_st_alpha).T

# Arredondamento dos valores de MAE para 3 casas decimais
list_MAE_st_alpha_T_arr = np.round(list_MAE_st_alpha_T,3)

# lista [10,1] com os nomes das colunas das tabelas 1 e 2
columns_names = alpha

# lista [52,1] com os indices das séries das tabelas 1 e 2
index_names = []
for i in range(1,len(list_MAE_st_alpha_T_arr)+1):
    index='st'+str(i)
    index_names.append(index)

# lista com os valores de MAE, valores de alpha e indices das séries temporais
MAE_result = [list_MAE_st_alpha_T_arr,columns_names, index_names]
```

### 2.2. Determinação da similaridade espacial (entre séries) baseada na matriz de correlação

$$ \mathbf{C_X} = \frac{1}{5760} (X - m)'(X - m)$$

$$ \mathbf{C_X}_{ij} = \frac{1}{5760} \sum_{k=1}^{5760} (X_{ik} - m_i)(X_{jk} - m_k) $$

```python
Xtr = Sens[0:5760]
N = Xtr.shape[0]

C = np.dot((Xtr - np.mean(Xtr, axis=0)).T,
                Xtr - np.mean(Xtr, axis=0)) / N
print(C.shape)
C
```

$$ \mathbf{Ro_X} = [ \mathbf{Ro_X}_{ij} = \frac{C_{ij}}{\sqrt{v_iv_j}} ]. $$

```python
var = np.var(Xtr, axis=0)
print(var.shape)
var
```

```python
#Ro
var = np.var(Xtr, axis=0) ##variancia das ST
list_st_Ro = []

for i in range (0,Xtr.shape[1]):
    list_Ro = []
    for j in range(0,Xtr.shape[1]):    
        Ro = (C[i,j] / np.sqrt(var[i]*var[j]))
        list_Ro.append(Ro)
    #lista [52,52] dos coeficiente de correlacao das ST
    list_st_Ro.append(list_Ro)

Ro = np.array(list_st_Ro)
Ro
```

```python
#similaridade entre ST utilizando C e Ro

def Get_Corr(mat_corr):
    corr_max = []
    corr_min = []
    
    for i in range(0, Xtr.shape[1]):
        list_corr_max = []
        list_corr_min = []
        
        st = list(mat_corr[i])
        st[i] = 0 # zera os elementos da diagonal principal da matriz de correlação
        index_max = np.argsort(st)[::-1] #cria uma matriz dos indices em ordem decrescente
        
        # As 3 mais similares
        list_corr_max.append(index_max[0]+1)
        list_corr_max.append(index_max[1]+1)
        list_corr_max.append(index_max[2]+1)
        
        # As 3 menos similares
        list_corr_min.append(index_max[50]+1)
        list_corr_min.append(index_max[49]+1)
        list_corr_min.append(index_max[48]+1)

        corr_max.append(list_corr_max)
        corr_min.append(list_corr_min)

    return corr_max, corr_min
```

```python
# Calculando as 3 series limilares entre usando o C
corr_max_C, corr_min_C = Get_Corr(C)
# Salvando dos dados das tabelas
np.savetxt('corr_max_C.txt', np.array(corr_max_C) , fmt='%4.f', delimiter='')
np.savetxt('corr_min_C.txt', np.array(corr_min_C), fmt='%4.f', delimiter='')
print('ST mais similares usando C: ' + str(corr_max_C[16])) #ST17
print('ST menos similares usando C: '+ str(corr_min_C[16])) #ST17
```

### 2.3. Imputação de dados ausentes utilizando matriz de correlação

Considere uma matriz $N \times M$ de séries temporais $Xtr = [st_{i,j}], i = 1, ..., M, j = 1, ..., N$, onde $M$ é o número de séries temporais, nesse caso $M = 52$, e $N$ é o número de pontos de cada série temporal, nesse caso $N = 5760$. Assim, $st_{i,j}$ representa o ponto da série temporal $st_{.,j}$ no tempo $i$. Cada coluna de $Xtr$ armazena uma série temporal.\
Os testes devem ser com o dataset de teste $Xte$. No dataset deste relatório não tem dados ausentes, assim a perda de dados será simulada em 10\% nos dados de testes. 

#### 2.3.1. Imputação com estimação espacial

```python
# Função para encontrar a matriz Xtr_i item 2.3.1
def Get_Xtri(Xtr,list_max_sim):
    
    Xtri = []
    # Percorrendo as 52 series temporais
    for i in range(0, Xtr.shape[1]):
        # Pegando as 3 mais similares de cada uma das 52 series
        max_st3 = list_max_sim[i]
        # Aqui vai as 4 series temporais [st.,i, st.,u, st.,v, st.,w]
        list_st_sim = []
        # Pgando a serie st.,i
        list_st_sim.append(Xtr[:,i])
        
        # Pegando as series st.,u, st.,v, st.,w
        for j in range(0, len(max_st3)):
            list_st_sim.append(Xtr[:,max_st3[j]-1])
        # Junatdo as lista Xtr_i
        Xtri.append(list_st_sim)
    
    return np.array(Xtri)

Xtri = Get_Xtri(Xtr, corr_max_Ro)
print(Xtri.shape)
print(Xtri[0].shape)
```

```python
# Função para obter o vetor m_i e matriz Cx_i
# [5760, 4]
def Get_Mi_Ci(Xtri):
    list_Cxi = []
    list_mi = []
    
    for i in range(0, Xtri.shape[0]):
        Stri =  Xtri[i].T # Transformando para [5760, 4]
        # Encontrado o vetor m_i da Xtr_i
        mi = np.mean(Stri, axis=0)
        # Criando a matriz Cx_i
        Stri_cen = Stri - mi
        Cxi = (1/Xtri.shape[2]) * \
                np.dot(Stri_cen.T, Stri_cen)
        # juntando matriz com todos os Cx_i
        list_Cxi.append(Cxi)
        list_mi.append(mi)
        
    return  np.array(list_Cxi), np.array(list_mi) # Retorna matriz com todos os Cx_i e m_i

Cxi, mi = Get_Mi_Ci(Xtri)
print(Cxi.shape)
print(mi.shape)
```

### Aplicando o teorema 12.7
$$
    \hat{X} = a'Y+b
$$$$
    Y = [X_1, X_2, ..., X_n]
$$$$
    a' = C_{X,Y}C_{Y}^{-1}
$$$$
    b = E[X]-C_{X,Y}C_{Y}^{-1}E[Y] = E[X] - a'E[Y]
$$

```python
def Get_Coef_Est(mat_cov, mat_m):
    est_x = []
    for i in range(0, 52): # Quantidade de series
        
        # Encontrado a o valor esperado E[Y] (Media)
        #E_Y = mat_m[i][1:] # as medias das tres series mais similares a i
        # Encontrado a matriz de covariancia de CY em C_X
        C_Y = mat_cov[i].copy()
        C_Y =  np.delete(C_Y, [0], axis=0) # removendo a linha de i
        C_Y =  np.delete(C_Y, [0], axis=1) # removendo a coluna de i
        # Encontrado a matriz de covariância cruzada C_x_y com Y = [X1, X2, X3]
        C_x_y = mat_cov[i][1:,0] # Pegando os valores cruzados C_x_y[1, 0], C_x_y[2, 0], C_x_y[3,0]
        # Calculando o coeficiente a'
        a = np.dot(C_x_y, np.linalg.inv(C_Y))
        # Calculando o coeficente b
        b = mat_m[i][0] - (a[0] * mat_m[i][1]) - (a[1] * mat_m[i][2]) - (a[2] * mat_m[i][3])
        
        # Estimando  o valor de X^
        #X = np.dot(a.T, E_Y) + b
        
        est_x.append([a, b])
    # Retorna os coefientes  de estimativa (a, b) de cada serie
    return est_x
```

```python
def Input_Data(data_test):
    #np.count_nonzero(np.isnan(data))
    # pegando os 10% de teste
    n = 0.10 * data_test.shape[0]
    rng = np.random.default_rng()
    # Lista para guarda a posicao dos dados imputados
    list_index = []
    # Imptando os dados
    data_input = data_test.T.copy()
    for i in range(0, 52):
        # Pegando 10% dos index de forma aleatoria de todas as series 
        indexs = rng.choice(range(0, data_test.shape[0]), size=(int(n)), replace=False)
        for j in range(0, int(n)):
            data_input[i][indexs[j]] = None      
        list_index.append(list(indexs))
    # Retorna os dados de teste com as imputacao e a posicao das imputacao
    return data_input, list_index
```

```python
# Criando os 10% de dados imputados aletoriamente
dados_input, posi_input = Input_Data(Xte)
```

```python
# Colculando os coeficientes ('a, b) para todas as series
coe_estima_x = Get_Coef_Est(Cxi, mi)
```

```python
# Estimativa espacial dos valores imputados
def EstimadorEspacial(coe_estima, dados_input, posi_input, sim_max3, dados_test):
    
    dados_test = dados_test.T # Transformnado para de 5760x52 para 52x5760
    
    for i in range(0, 52):
        
        a = coe_estima[i][0] # Coeficiente a da serie i
        b = coe_estima[i][1] # Coeficiente b ca serie i
        
        st_max3 = [] # as tres series mais similares a serie i
        index_input = posi_input[i] # Os index de onde foram imputados os dados da serie i
        # Pegando as 3 series mais similares a i nos dados de teste
        st_max3.append(dados_test[sim_max3[i][0]-1])
        st_max3.append(dados_test[sim_max3[i][1]-1])
        st_max3.append(dados_test[sim_max3[i][2]-1])
        
        # Estimar todos os dados imputados da serie i
        for j in range(0, len(index_input)):
            # Vetor Y
            Y = []
            # Pegando os 3 valores das series mis similares onde esta faltando dados em i
            #print(max3[0], index_imput[j])
            Y.append(st_max3[0][index_input[j]])
            Y.append(st_max3[1][index_input[j]])
            Y.append(st_max3[2][index_input[j]])
            
            # Estimando o valor cm os coeficentes da serie i
            # Aplicando a formula
            Xi = np.dot(a.T, Y) + b
            
            # Substituido os dados faltosos pela estimativa 
            dados_input[i][index_input[j]] = Xi
            
    # Retornando os dados de teste com os valores imputados substituidos        
    return dados_input

# Estimando os novos dados de testes pela tecnica de estimativa espacial
esti_dados_test_esp = EstimadorEspacial(coe_estima_x, dados_input, posi_input, corr_max_Ro, Xte)
esti_dados_test_esp.shape
```

```python
# Estimativa por media dos dados de treinamento
# Estimativa por valor anterior
def estiMean(dados_input, index_input, vetor_medias):
    dados_esti_mean = dados_input.copy()
    dados_esti_valor = dados_input.copy()
    
    for i in range(0, 52):
        # A media da serie i 
        media_sti = vetor_medias[i][0]
        # Serie i com valores faltosos
        sti_input = dados_input[i]
        # Index das posicoes dos dados faltosos de i
        st_index = sorted(index_input[i])
        
        for j in range(0, len(st_index)):
            # Substituindo os dados pela media
            dados_esti_mean[i][st_index[j]] = media_sti
            
            # Substituindo pelo valor anterior
            # Varificando se a posicao é 0 e se o valor anterior nao faltoso
            if st_index[j] == 0 and np.isnan(dados_esti_valor[i][st_index[j]+1]) == False :
                
                dados_esti_valor[i][st_index[j]] = dados_esti_valor[i][st_index[j]+1]
            
            # se o valor faltoso for na posicao 0 da serie
            elif st_index[j] > 0:
                dados_esti_valor[i][st_index[j]] = dados_esti_valor[i][st_index[j]-1]
            
            else:
                dados_esti_valor[i][st_index[j]] = dados_esti_valor[i][st_index[j]+2]
    
    return dados_esti_valor, dados_esti_mean
```

#### 2.3.2. Imputação com estimação temporal
Na imputação com estimação temporal repete-se o mesmo procedimento da subseção 2.3.1 sendo a diferença em como $Xtr_i$ é obtida. Neste caso, $Xtr_i$ é formada por $st_{i,.}$ e versões de $st_{i,.}$ atrasadas no tempo, ou seja, $Xtr_j = [st_{i,j}, st_{i-1,j}, st_{i-2,j}, st_{i-3,j}]$. Todo o restante do procedimento é o mesmo da subseção 2.3.1.

```python
# Criando a matriz Xtrj
#i = 0..., 5760
#j = 0..., 52
#Xtrj = Xtr[:,j][i-1]

Xtrj = []
for j in range(0, 52): # As 52 series
    stj = Xtr[:,j]
    st_atra_3 = []
    st_atra_3.append(stj)
    # Os tres valores de atraso
    for k in range(1, 4): #  k = 1, 2, 3
        stj_atra = []
        # 0 a 5760, percorrendo todos os valores da serie
        for i in range(0, stj.shape[0]):
            stj_atra.append(stj[i-k])
        # Pegando a serie atrasada com -1, -2, -3
        st_atra_3.append(stj_atra)
    # Adicionando a serie j e as tres atrasadas
    Xtrj.append(st_atra_3)
```

```python
def Get_Xtrj(Xtr):
    
    Xtrj = []
    for j in range(0, 52): # As 52 series
        stj = Xtr[:,j]
        st_atra_3 = []
        st_atra_3.append(stj)
        # Os tres valores de atraso
        for k in range(1, 4): #  k = 1, 2, 3
            stj_atra = []
            # 0 a 5760, percorrendo todos os valores da serie
            for i in range(0, stj.shape[0]):
                stj_atra.append(stj[i-k])
            # Pegando a serie atrasada com -1, -2, -3
            st_atra_3.append(stj_atra)
        # Adicionando a serie j e as tres atrasadas
        Xtrj.append(st_atra_3)
    
    return np.array(Xtrj)
```

```python
# Estimativa espacial dos valores imputados
def EstimadorTemporal(coe_estima, dados_input, posi_input, dados_test):
    
    dados_test = dados_test.T # Transformnado para de 5760x52 para 52x5760
    dados_input_ = dados_input.copy()
    for i in range(0, 52):
        
        a = coe_estima[i][0] # Coeficiente a 1x3 da serie i
        b = coe_estima[i][1] # Coeficiente b 1x1 ca serie i
        
        index_input = sorted(posi_input[i]) # Os index de onde foram imputados os dados da serie i
        # Pegando as 3 series mais similares a i nos dados de teste
        
        # Estimar todos os dados imputados da serie i
        for j in range(0, len(index_input)):
            # Vetor Y
            Y = []
            # repetindo os valores para a comtibilidade com os coeficentes espaco-temporal      
            
            Y.append(dados_input_[i][index_input[j]-1])
            Y.append(dados_input_[i][index_input[j]-2])
            Y.append(dados_input_[i][index_input[j]-3])
            
            # Estimando o valor cm os coeficentes da serie i
            # Aplicando a formula
            X_i = np.dot(a.T, Y) + b
            
            # Substituido os dados faltosos pela estimativa 
            dados_input_[i][index_input[j]] = X_i
            
    # Retornando os dados de teste com os valores imputados substituidos        
    return dados_input
```

```python
# Estimendo os valores amputados
def EstimadorEspacialTemporal(coe_estima, posi_imput, index_sim_max3, dados_test):
    dados_test = dados_test.T # Transformnado para de 5760x52 para 52x5760
    novos_dados = dados_test.copy()
    for i in range(0, 52):
        
        a = coe_estima[i][0] # Coeficiente a 1x3 da serie i
        b = coe_estima[i][1] # Coeficiente b 1x1 ca serie i
        index_imput = sorted(posi_imput[i]) 
        # Estimar todos os dados imputados da serie i
        for j in posi_imput[i]:
            Y = []
            # Pegando os valores das similares
            Y.append(dados_test[index_sim_max3[i][0]-1][j])
            Y.append(dados_test[index_sim_max3[i][0]-1][j])
            Y.append(dados_test[index_sim_max3[i][0]-1][j])
            # Pegando os valores atrazados
            Y.append(dados_test[i][j-1])
            Y.append(dados_test[i][j-2])
            Y.append(dados_test[i][j-3])
            # Estimando o valor cm os coeficentes da serie i
            X_i = np.dot(a.T, Y) + b
            # Substituido os dados faltosos pela estimativa 
            novos_dados[i][j] = X_i
    # Retornando os dados de teste com os valores imputados substituidos        
    return novos_dados
```
